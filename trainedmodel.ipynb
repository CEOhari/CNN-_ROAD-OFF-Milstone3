```python
import matplotlib.pyplot as plt
from keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np 
```

```python
#load trained model 

model = load_model('my_model.keras')
```

```python
from tensorflow.keras.preprocessing.image import load_img

```

```python
# Load the image
image_path = r"D:\Road Signs classification dataset\Dataset\Speedlimit_15\001_0003.png"
image = load_img(image_path, target_size=(160, 160))
```

```python
#preprocess the image 
img = np.array(image)
img = img / 255.0
img = img.reshape(1,160,160,3)
```

```python
#predict the label
label = model.predict(img)
```

```python
class_names = ['5_miles_speedlimit', 'No_entry', 'Speedlimit_15']
predicted_class_index = np.argmax(label)
predicted_class = class_names[predicted_class_index]

print("Predicted Class:", predicted_class)
```

```python
print("predicted_class:", predicted_class)
```

```python
plt.imshow(image)
plt.title(predicted_class)
plt.axis('off')
plt.show()
```
